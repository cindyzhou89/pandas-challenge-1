## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
df['unit_price_sum'] = df['unit_price']*df['qty']
df[['unit_price','qty','unit_price_sum']].head(2)
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound
# for orders over 50 pounds and $10 per
# pound for items 50 pounds or under.
total_weight = df['unit_weight'] * df['qty']
df['total_weight']=total_weight
def cal_shipping_rate(weight):
    if weight >50:
        return weight * 7
    else :
        return weight * 10
df['shipping_price'] = df['total_weight'].apply(cal_shipping_rate)
df[['unit_price','unit_weight','qty','total_weight','shipping_price']].head(3)
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
#df['subtotal']=df['unit_price']*df['qty']
sales_tax = 0.0925
df['total_price']= (df['unit_price_sum']+df['shipping_price'])*(1+sales_tax)
df['total_price']= df['total_price'].round() 
df[['unit_price','unit_weight','qty','total_weight','shipping_price','total_price']].head(3)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
df['cost']=df['unit_cost']*df['qty']+df['shipping_price']
df[['cost','unit_cost','qty','shipping_price']].head(3)
#del df['subtotal']
```

```python
# Create a column for the profit of
# each line using line cost and line price
df['profit'] = df['total_price'] - df['cost']

df[['profit','total_price','cost']].head(3)
```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
client_vips_df = pd.DataFrame(client_vips)

client_vips_df = client_vips_df.sort_values(by='line_profit')
client_vips_df
 
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.

desc_profit_vip=client_vips_df.rename({'client_id': 'Client ID', 'qty': 'Units','shipping_price': 'Shipping', 
                 'line_price': 'Total Revenue','line_cost': 'Total Cost', 'line_profit': 'Total Profit'}, axis=1)
format_coloumns = ['Shipping', 'Total Revenue', 'Total Cost', 'Total Profit']
desc_profit_vip_overview = desc_profit_vip.copy()

# Apply the formatting to the specific columns
desc_profit_vip_overview[format_coloumns] = desc_profit_vip_overview[format_coloumns].apply(lambda x: x.apply(lambda val: f'${val / 1_000_000:.2f}M'))

```

```python
# Sort the updated data by "Total Profit" form highest to lowest

desc_profit_vip_overview=desc_profit_vip_overview.sort_values(by='Total Profit', ascending=False)
desc_profit_vip_overview
```
